```python
import pandas as pd
import datetime
import numpy as np
```

```python
data = pd.read_csv("../csv/scores_without_news.csv")
```

```python
data.set_index("date", inplace=True)
```

```python
data
```

```python
test = data.copy()
```

```python
test.set_index("date", inplace = True)
```

```python
def replace_values(cols,value, new_value):
    for col in cols:
        test[col] = test[col].replace(value, new_value)
```

```python
replace_values(["ma","axp","cof"], "positive", 1)
replace_values(["ma","axp","cof"], "negative", -1)
replace_values(["ma","axp","cof"], "neutral", 0)
```

```python
test
```

```python
test["date"] = pd.to_datetime(test["date"])
```

```python
test = test.resample("D").agg({
    "ma": "mean",
    "axp": "mean",
    "cof": "mean",
    "news_score":"mean"
})
```
