```python
def es_ingles(text):
    if detect(text) == "en":
        return True
    else:
        return False
```

```python
data2 = test.copy()
```

```python
data2
```

```python
data2["es_ingles"] = data2.title.apply(es_ingles) 
```

```python
notiocias_no_ingles = data2.loc[data2.es_ingles != True]
```

```python
notiocias_no_ingles
```

```python
noticias_prueba2 = pd.DataFrame(notiocias_no_ingles)
```

```python
noticias_prueba2["es_ingles"] = noticias_prueba2.title.apply(es_ingles) 
```

```python
noticias_prueba2
```

```python
noticias_falso_negativo_ingles = noticias_prueba2.iloc[27:]
```
