```python
import pandas as pd
```

```python
import os as os 
```

```python
import requests
```

```python
missing_dates = pd.read_csv("./data/fechas_faltantes.csv")
```

```python
missing_dates
```

```python
missing_dates.drop(columns="Unnamed: 0", inplace= True)
```

```python
missing_dates.rename(columns={"0":"dates"}, inplace= True)
```

```python
missing_dates
```

```python
prepared_dates = list(map(lambda date: str(date)[:10],list(missing_dates.dates))) 
```

```python
prepared_dates
```

```python
api_key = os.getenv("API_KEY")
```

```python
def get_config(date):
    return{
        "ticker": "MA",
        "published_utc": date,
        "limit":"5",
        "apiKey":api_key
    }
    
```

```python
getted_news = []
```

```python
import time
```

```python
for date in prepared_dates:
    try: 
        getted_news.append(requests.get("https://api.massive.com/v2/reference/news", params= get_config(date)))
        time.sleep(2)
    except Exception as e:
        print(f"no se pudo hacer esto, ya que: {e}")
```

```python
print([new.json() for new in getted_news])
```
