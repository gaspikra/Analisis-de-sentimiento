```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('../data/csv/final.csv')
```

```python
data.drop(['Date'], axis=1, inplace = True)
```

```python
import seaborn as sns
```

```python
corr = data.corr()
```

```python
axs = sns.heatmap(
    corr,
    vmin=-1,
    vmax=1,
    annot=True,
    fmt='.1f',       
    cmap='coolwarm', 
    linewidths=.5)
plt.show()
```

```python
# para borrar: Open, high, low, close, adj_close, stock splits, dividends
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=data['news_score'])
plt.title('Detección de Outliers en Sentiment')
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))
#precio
ax1.plot(data.index, data['Close'], color='blue', label='Precio')
ax1.set_ylabel('Precio', color='blue')

#sentimiento
ax2 = ax1.twinx()
ax2.fill_between(data.index, data['news_score'], color='orange', alpha=0.3, label='Sentimiento')
ax2.set_ylabel('Sentimiento', color='orange')

plt.title('Relacion Temporal: Precio vs Sentimiento')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.regplot(data=data, x='news_score', y='Close', 
            scatter_kws={'alpha':0.5, 'color':'teal'}, 
            line_kws={'color':'red'})

plt.title('Dispersion: Sentimiento vs Precio de Cierre')
plt.xlabel('Sentimiento (Noticias)')
plt.ylabel('Precio de Cierre (USD)')
plt.grid(True, alpha=0.3)
plt.show()
```
